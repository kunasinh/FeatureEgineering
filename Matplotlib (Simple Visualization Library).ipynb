```
# compute the x and y co ordinates for points on a sine curve

x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)

plt.title("Sine Wave Form")

# plotting the graph

plt.plot(x,y)
```

```
# compute the x and y co ordinates
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#set up a subplot grid
plt.subplot(2,1,1)

#make the first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine')

# second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')
```

```
# Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2, color = 'r')

plt.title('Bar Graphs')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```
# Histograms
# it shows the density of the values present in the range

a = np.array([22,21,18,3,3,30,23,47,44,5,34,34,4,34,33,77,86,90])
plt.hist(a)
plt.title('Histogram')
```

```
# Box Plot
# it helps in finding percentiles

data = [np.random.normal(0,std, 100) for std in range(1,4)]

# rectangular box plot

plt.boxplot(data, vert = True, patch_artist=True)

# the circles in plot are outliers
```

```
# Pie Chart

# Data to PLot

labels = 'Pyhton', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.2, 0,0,0)

#plot

plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow= True)

plt.axis('equal')
```
